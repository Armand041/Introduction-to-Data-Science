There are 3 main **Learning Paradigms** in actual **Machine Learning**: **Supervised Leraning**, **Unsupervised Learning** and **Reinforced Learning**. In this section we will talk about the first two paradigms, but we will view them as 'black boxes'. In other words, we will learn what they are capable of and their characteristics without entering in the details of how they do it.

## 3.1. Supervised Learning

**Supervised Learning** is one of the most common types of learning paradigms, but it often requires the most resources to work and function correctly, as it normally needs human interaction to iniate its learning process.

### 3.1.1. Theory

**Supervised Learning** finds patterns or relationships between pairs of inputs and outputs. It is called supervised as, normally, human interaction is needed to train the algorithm. The task of the human trainer, from now on the **teacher**, will be to feed the algorithm inputs and their respective outputs. This initial batch of information will help the AI to learn the relationship between them, enabling it to generalize the knowledge and use it in new, neverseen, inputs.

Let us see an example to better understand the concept:

|Input|Output|
|-|-|
|1|1|
|2|4|
|3|9|
|6|36|
|8|64|
|11|121|
|20|400|

Given this initial batch, we, as humans, can clearly see a pattern pop up. The output is the input multiplied by itself (its square number). So, if new input arrived such as the number 7, we could apply the generalization we just made and provide the output 49.

In a more programatic way, we could say the following:

```python
def generalization(n: int) -> int:
    '''
    Given an integer input, return the squared input

    n: input
        -> Integer number that represents the input of a AI using supervised learning
    '''

    return n**2
```

```python
print(f'Input: 7 -> Output: {generalization(7)}')
print(f'Input: 15 -> Output: {generalization(15)}')
print(f'Input: 25 -> Output: {generalization(25)}')
print(f'Input: 100 -> Output: {generalization(100)}')
```

This can be applied to any input-output situation, wether e-mails are spam or not, if a student is going to pass or fail exams based on their notes, what is the favourite candy of a kid given its favourite sport... If there is any pattern or relationship between the input and the output (and given enough examples by the teacher), the algorithm will find it.

It is a heavily used learning paradigm, and for good reasons, but, as we saw, it comes with a hefty price. It requires a teacher to manually process some quantity of the data, normally around 20% of it. The more data we manually process, the better the algorithm will perform, so it is a question of how much time and labour are we willing to invest in the learning batch of data and what precision of the output we desired.

## 3.2. Unsupervised Learning

Contrary to **supervised learning**, it does not require an initial batch of input-output examples to function. It is this characteristic that makes this language paradigm essential in the future of **machine learning**.

### 3.2.1. Theory

**Unsupervised Learning** finds the internal structure of the inputs, creating and forming patterns of similitude between the inputs. The more complex algorithms can determine the underlying structure and characteristics of the inputs, creating what is called a **Latent Space**. This way they are capable of generating and learning about similar inputs, creating a complex network of relations that allows us to mathematically operate them.

One of the most common problems of **Unsupervised Learning** are the **Clustering** problems. This problems study the optimal way to divide in groups a set of inputs. Let us see a **clustering** example to visualize this learning paradigm:

Input -> a, b, 1, 5, c, e, 8, f, 6, 4

Option 1, dividing in letters and numbers:

    First Cluster: a, b, c, e, f

    Second Cluster: 1, 5, 8, 6, 4

Option 2, dividing in vowels, consonants, evens and odds

    First CLuster: a, e

    Second Cluster: b, c, f

    Third Cluster: 8, 6, 4

    Fourth CLuster: 1, 5

Having more **clusters** makes for a more concrete division, however, if we keep increasing them we may run into an algorithm too individualistic to form meaningful connections. This would waste resources and defeat the focus of **clustering** problems. Studying the optimal number of **clusters** is essential for a good algorithm.

This way of learning is much cheaper than **Supervised Learning**, but the results are impossible to verify, as it does not follow an input-output basis. This type of algorithm is essential in generative AI, as it mathematically approaches the inputs, creating **latent spaces** that organize the information by similarities in their underlying structure. Substracting, adding and multiplying the vectors inside this mathematical space transforms the input into the output.

## 4.1. Models and Probabilistic Models

To understand our complex world, we make use of simplified versions of it. Diagrams, musical sheets, maps... they all are models to a more difficult to understand reality. The objective of a model is to balance the precision when representing reality with enough simplicity to be effective. 

For example, the maps lack a 100% precision when representing reality, as the depth, colors and textures are often lost in a 2D map. However, this imprecission is acceptable for the sake of simplicity and effectivity when using it.



When we start to model real life, we might encounter with a problem: too much complexity. We can try to model students passing exams, but we rapidly encounter exceptions upon exceptions. The model could be the next one:

'A student passes the exam if they have...

    -> studied
    
    -> slept well
    
    -> have a fondness for the subject
    
    -> has not received negative news in the past week
    
    -> copied and not get caught
    
    -> ...'

The list can go on and on, adding exceptions (or **Parameters** as we will learn in the next section) to try reaching the 100% reality precision, but the complexity of the model will also increase.  This creates a flawed way of modelling, as too much complexity can render useless any effort of computing or creating the models. Instead, we can use **Probabilistic Models** as a short-cut to reduce complexity whilst maintaining a relatively high fidelity to reality.

Now, instead of saying a long list of requirements to pass the exam, we can simply say the probability of passing it:

'A student has an 85% probability of passing the exam'

They can also be combined, having the probability tied with some important exceptions such as the following example:

'A student that has...

    -> studied
    
    -> slept well

has an 95% of probability of passing the test'

Both ways of modelling have their strengths and weaknesses, so each one can come in handy when reviewing certain parts of real life. However, they share some objectives and characteristics that we will study and review in the next parts: **Error Function** and **Error Optimization**.

## 4.2. Data in Models

When talking about **Models**, it is essential to talk about data and parameters within the reach of the model. As we saw in the previous section, we can feed the model with exceptions, or, to be more precise, **Parameters**. 

We will start by creating a simple model to understand the importance of **Parameters** and **Multidimensionality**. The model will be the students passing an exam based on their study time, in other words, our parameters will be study time and passing the exam.

```python
import matplotlib.pyplot as plt

# Study time in hours
study_time = [1, 2, 3, 5, 7, 8.5]
# Passed the exam (1) or not (0)
passed = [0, 0, 0, 0, 1, 1]

# Create the frame
plt.figure(figsize =(10,2.5))

# Plot the data
plt.scatter(study_time, passed)

# Set the axis labels
plt.xlabel('Study Time (h)')
plt.ylabel('Passed (1/0)')

# Show the plot
plt.show()
```

We can represent the two parameters (Passed and Study Time) in a 2 dimensional graph. We can adjust this parameters to better fit the reality, however, this may fall of short. If this happens, we need to increase the number of parameters, increasing the number of dimensions needed to represent them. 

Let us run again the same model, but adding a third parameter: Time Slept.

```python
import matplotlib.pyplot as plt

# Study time in hours
study_time = [1, 2, 3, 5, 7, 8.5]
# Passed the exam (1) or not (0)
passed = [0, 1, 0, 1, 1, 0]
# Slept Time in hours
time_slept = [8, 7, 5, 8, 6.5, 3]

# Create the frame
fig = plt.figure(figsize =(5,5))
ax = fig.add_subplot(111, projection= '3d')

# Plot the data
ax.scatter(passed, study_time, time_slept)

# Set the axis labels
ax.set_xlabel('Passed (1/0)')
ax.set_ylabel('Study Time (h)')
ax.set_zlabel('Time Slept (h)', labelpad=-3)

# Show the plot
plt.show()
```

From this point on, adding parameters will only increase the **dimensionality** of the data, making it harder and harder to visualize. However, mathematically speaking, adding dimensions does little impact in the complexity. From this point on, we will see the data as points on a vector map, with as many dimensions as **parameters**.
